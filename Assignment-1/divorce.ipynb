```python
comparison_result = df["income_man"]<= df["income_woman"]
#To select the values where

print(comparison_result)

true_values_df = df[comparison_result == True]
false_values_df = df[comparison_result == False]
#To select the values that are true and false separately

print(true_values_df)
print(false_values_df)

true_values_count = len(true_values_df)
false_values_count = len(false_values_df)
#To calculate the number of true values and the number of false values

print(true_values_count)
print(false_values_count)
```

```python
categories = ['Women > Men', 'Men > Women']
counts = [true_values_count, false_values_count]

plt.barh(categories, counts, color=['blue', 'red'])

plt.xlabel('Number of Cases')
plt.title('Comparison of Salaries in Divorce Cases')

plt.show()
```
