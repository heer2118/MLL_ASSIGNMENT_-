```python
mean_unemployment = df[['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021']].mean(axis=0)

plt.plot(mean_unemployment.index, mean_unemployment.values)

#set labels and title
plt.xlabel('Year')
plt.ylabel('Mean Unemployment Rate')
plt.title('Line Graph of Mean Unemployment Rate from 2010 to 2021')
plt.show()
```
